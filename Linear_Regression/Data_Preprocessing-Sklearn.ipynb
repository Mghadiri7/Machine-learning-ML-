```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import joblib
from sklearn import preprocessing
```

```python
df = pd.read_csv("student-mat.csv", sep=";")
df = df[["sex", "absences", "school", "Medu", "G1", "G2", "G3"]]
predict = "G3"
```

```python
myPreprocessing = preprocessing.LabelEncoder()
sex = myPreprocessing.fit_transform(list(df["sex"]))
school = myPreprocessing.fit_transform(list(df["school"]))
```

```python
#Example Zip
lst1 = [1,2,3,4,5]
lst2 = ["A","B","C","D","E"]
lst3 = ["@","#","$","%","&"]
print(list(zip(lst1,lst2,lst3)))
```

```python
X = list(zip(df["absences"], df["Medu"], df["G1"], df["G2"], sex, school))
```

```python
X
```

```python
y = np.array(df[predict])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)
```
